```python
import os

from spell_checking import BENCHMARK_DIR
from nsc.api import utils
```

```python
spelling_correction_neuspell = os.path.join(BENCHMARK_DIR, "test", "sec", "spelling_correction", "neuspell")
```

```python
corrupt = utils.load_text_file(os.path.join(spelling_correction_neuspell, "corrupt.txt"))
correct = utils.load_text_file(os.path.join(spelling_correction_neuspell, "correct.txt"))

misspelled_lines_and_words = set()
for i, (corrupt_l, correct_l) in enumerate(zip(corrupt, correct)):
    for j, (corrupt_w, correct_w) in enumerate(zip(corrupt_l.split(), correct_l.split())):
        if corrupt_w != correct_w:
            misspelled_lines_and_words.add((i, j))
len(misspelled_lines_and_words)
```

```python
predictions = utils.load_text_file(os.path.join(BENCHMARK_DIR, "test", "sec", "results", "spelling_correction", "neuspell", 
                                                "transformer_sec_words_nmt.txt"))
beam_predictions = utils.load_text_file(os.path.join(BENCHMARK_DIR, "test", "sec", "results", "spelling_correction", "neuspell", 
                                                "transformer_sec_words_nmt_beam.txt"))
```

```python
for i, (pred_l, pred_beam_l) in enumerate(zip(predictions, beam_predictions)):
    for j, (pred_w, pred_beam_w) in enumerate(zip(pred_l.split(), pred_beam_l.split())):
        if pred_w != pred_beam_w and (i, j) in misspelled_lines_and_words:
            print("corrupt:", corrupt[i], "correct:", correct[i], "pred:", pred_l, "pred beam:", pred_beam_l, "-" * 30, sep="\n")
```
